```python
#inmport library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
#Load the Data
data = pd.read_csv('Documents/application_train.csv')
data.head()
```

```python
#EXPLORATORY DATA ANALYSIS
data.info('column')
```

```python
korelasi = data.select_dtypes(include=['float64', 'int64']).columns
corr_pearson = data[korelasi].corr(method='pearson')
print(corr_pearson)
```

```python
#DROP DATA
db = data[['SK_ID_CURR','TARGET','NAME_INCOME_TYPE','REGION_POPULATION_RELATIVE','OCCUPATION_TYPE','ORGANIZATION_TYPE','EXT_SOURCE_2','EXT_SOURCE_3','DAYS_BIRTH','DAYS_EMPLOYED','AMT_CREDIT','AMT_ANNUITY',
          'NAME_CONTRACT_TYPE','AMT_INCOME_TOTAL','AMT_GOODS_PRICE','NAME_EDUCATION_TYPE','NAME_FAMILY_STATUS','FLAG_MOBIL','FLAG_EMAIL','DAYS_REGISTRATION','DAYS_ID_PUBLISH']]
print(db)
```

```python
db.isnull().sum()
```

```python
#Preprocessing Data
db['NAME_INCOME_TYPE'] = db['NAME_INCOME_TYPE'].map({'Working':1, 'State servant':2,'Pensioner':3,'Commercial associate':4})
db['NAME_FAMILY_STATUS'] = db['NAME_FAMILY_STATUS'].map({'Married':1, 'Single / not married':2,'Civil marriage':3,'Widow':4})
db['NAME_EDUCATION_TYPE'] = db['NAME_EDUCATION_TYPE'].map({'Higher education':1, 'Secondary / secondary special':2})
db['NAME_CONTRACT_TYPE'] = db['NAME_CONTRACT_TYPE'].map({'Cash loans':1, 'Revolving loans':2})
db['OCCUPATION_TYPE'] = db['OCCUPATION_TYPE'].map({'Low-skill Laborers':1, 'Drivers':2,'High skill tech staff':3,'Core staff':4,'Laborers':5,'Sales staff':6,'Managers':7})
db['ORGANIZATION_TYPE'] = db['ORGANIZATION_TYPE'].map({'Kindergarten':1, 'Self-employed':2,'Transport: type 3':3,'Business Entity Type 3':4,'Government':5,'Industry: type 9':6,'School':7,'Trade: type 2':8,'Services':9,'Bank':10,'XNA':11})
db.head()
```

```python
db.describe().T
```

```python
df=db.fillna(db.mean())
df.head()
```

```python
df.isnull().sum()
```

```python
# Separing variables from values

X = df.drop(['TARGET'], axis = 1)#independent Variable
Y = df['TARGET']#dependent Variable

X.head()
```

```python
#find how many examples are there for 1 and 0
df['TARGET'].value_counts() #20 is the column index
```

```python
# Calling the Standardization library
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df.drop('TARGET',axis = 1))
```

```python
# Standardization Process
scaled_features = scaler.transform(df.drop('TARGET',axis = 1))
df_feat = pd.DataFrame(scaled_features,columns = df.columns[:-1])

df_feat.head()
```

```python
# Determine the training data and test data with a simple random sampling technique
X = df_feat
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Calling the LogisticRegression library
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(class_weight='balanced')
logmodel.fit(X_train,y_train)

# Prediction X_test for LogisticRegression
y_predict = logmodel.predict(X_test)

Predict = pd.DataFrame(np.array(y_predict),
                   columns=['Prediction'])
verbal=Predict['Prediction'].map({0:'Baik', 1:'Buruk'})
Predict['Conclusion'] = pd.DataFrame(np.array(verbal), 
                    columns=['Conclusion'])
Predict.head(100)
```

```python
# Report of Model Evaluation
print(f"Report Model Evaluation: \n {classification_report(y_test, y_predict)}")
print(f"Confusion Matrix: \n {confusion_matrix(y_test, y_predict)}")
```

```python
from sklearn.metrics import  roc_auc_score
roc_auc_score(y_test, y_predict)
```

```python
Predict.value_counts()
```

```python
# Calling the RandomForestClassifier library
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators =40, random_state = 0, class_weight='balanced')
forest.fit(X_train,y_train)  

# Prediction X_test for RandomForestClassifier
y_predic = forest.predict(X_test)

Predic = pd.DataFrame(np.array(y_predic),
                   columns=['Prediction'])
verbal=Predic['Prediction'].map({0:'Baik', 1:'Buruk'})
Predic['Conclusion'] = pd.DataFrame(np.array(verbal), 
                    columns=['Conclusion'])
Predic.head()
```

```python
# Report of Model Evaluation
print(f"Report Model Evaluation: \n {classification_report(y_test, y_predic)}")
print(f"Confusion Matrix: \n {confusion_matrix(y_test, y_predic)}")
```
